Cleaning

```python
#Taking bacckup of data
backup=car.copy()
```

```python
#keeping only valid year value
car =car[car['year'].str.isnumeric()]
#changing data typ of year from object to int
car['year']=car['year'].astype(int)
```

```python
car.info()
```

```python
car =car[car['Price'] != "Ask For Price"]
```

```python
#removing commans nd data type changing
car['Price']=car['Price'].str.replace(',' , '')
#Conversion of data type
car['Price']=car.Price.astype(int)
```

```python
#removing kms
#car['kms_driven']

#spliting string ,get first part and removing commans
car['kms_driven']=car['kms_driven'].str.split(' ').str.get(0).str.replace(',', '')

#removing string(not valid kms number)
car=car[car['kms_driven'].str.isnumeric()]
#data type change
car['kms_driven'] = car.kms_driven.astype(int)
```

```python
#keep first 3 value
car['name']=car['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
#reset the indexing
car=car.reset_index(drop='true')
#car.describe()
```

```python
#car[car['Price']>6e6] 
#it's outliear 
car=car[car['Price']<6e6].reset_index(drop='true')
```

```python
car.to_csv('Cleaned car.csv')
```

Model Traning

```python
x=car.drop(columns='Price')
y=car['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test =train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder 
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(x[['name','company','fuel_type']])
```

```python
coulmn_trans= make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']), remainder='passthrough')
```

```python
lr= LinearRegression()
```

```python
pipe= make_pipeline(coulmn_trans,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
#Predicting y values
y_pred=pipe.predict(x_test)
```

```python
#finding r2 score
r2_score(y_test,y_pred)
```

Finding the best value of  random state for highest value of r2_score

```python
from pandas.core.common import random_state
import numpy as np

score=[]
for i in range(1000):
  x_train, x_test, y_train,y_test =train_test_split(x,y,test_size=0.2,random_state=i)
  lr= LinearRegression()
  pipe= make_pipeline(coulmn_trans,lr)
  pipe= make_pipeline(coulmn_trans,lr)
  pipe.fit(x_train,y_train)
  y_pred=pipe.predict(x_test)
  #print(r2_score(y_test,y_pred) ,i)
  #Store all of r2_score in score list
  score.append(r2_score(y_test,y_pred))


```

```python
#Best value of random state for highest value of r2_score
np.argmax(score)
```

```python
#Highest value of r2_score
score[np.argmax(score)]
```

```python
#Train the model for best value of random_state
x_train, x_test, y_train,y_test =train_test_split(x,y,test_size=0.2,random_state=np.argmax(score))
lr= LinearRegression()
pipe= make_pipeline(coulmn_trans,lr)
pipe= make_pipeline(coulmn_trans,lr)
pipe.fit(x_train,y_train)
y_pred=pipe.predict(x_test)
r2_score(y_test,y_pred)


```
